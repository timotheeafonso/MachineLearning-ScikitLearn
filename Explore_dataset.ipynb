## California housing 

```python
housing_data = pd.read_csv('datasets/housing.csv')
```

```python
housing_data.head()
```

```python
housing_data.sample(5) # 5 record random
```

```python
housing_data= housing_data.dropna() # Drop record with missing values
```

```python
housing_data.describe() # Statistic information about all column with numeric values
```

```python
housing_data['ocean_proximity'].unique()
```

```python
# Tableau nombre de piece en fonction du prix de la maison
fig,ax = plt.subplots(figsize=(12,8))
plt.scatter(housing_data['total_rooms'],housing_data['median_house_value'])
plt.xlabel('Total rooms')
plt.ylabel('Median house value')
```

```python
# Tableau age de la maison en fonction du prix de la maison
fig,ax = plt.subplots(figsize=(12,8))
plt.scatter(housing_data['housing_median_age'],housing_data['median_house_value'])
plt.xlabel('Median age')
plt.ylabel('Median house value')
```

```python
# Tableau nombre de voisin en fonction du prix de la maison
fig,ax = plt.subplots(figsize=(12,8))
plt.scatter(housing_data['median_income'] * 10000,housing_data['median_house_value'])
plt.xlabel('Median income')
plt.ylabel('Median house value')
```

```python
# Affiche la correlation entre chaque  paire d'attribut du dataset
housing_data_corr = housing_data.corr()
housing_data_corr
```

```python
# Meilleur visualisation des correlation avec seaborn
fig,ax = plt.subplots(figsize=(12,10))
sns.heatmap(housing_data_corr,annot=True)
```
